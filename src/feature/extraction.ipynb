```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import pathlib
```

```python
x, sr = librosa.load('../../trackfiles/Funny Scammer Pranks LIVE  398  Dec 12 2020/Funny0.mp3')
```

```python
# hop_length = 10
n_fft = 512
X = librosa.stft(x,n_fft=n_fft)
```

```python
S = librosa.amplitude_to_db(abs(X))
```

```python
plt.figure(figsize=(15,5))
librosa.display.specshow(S,sr=sr,x_axis='time',y_axis='log')
plt.colorbar(format='%+2.0f dB')
```

```python
m_fcc = librosa.feature.mfcc(x,sr=sr)
```

```python
m_fcc = librosa.power_to_db(abs(S))
```

```python
plt.figure(figsize=(15,5))
librosa.display.specshow(m_fcc,x_axis='time',y_axis='hz')
plt.colorbar(format='%+2.0f dB')
```
